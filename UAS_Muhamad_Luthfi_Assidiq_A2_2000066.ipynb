```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from matplotlib.ticker import PercentFormatter
```

# ***Scatter Plot 1*** Kebutuhan di Tiap Propinsi Setiap Tahunnya

```python
# memanggil dataset kebutuhan per bulan 2
df = pd.read_csv('/content/kebutuhan_per_bulan_2.csv')
```

```python
df
```

```python
plt.figure(figsize=(12, 10))
plt.scatter(df['tahun'], df['Propinsi'], s=50, c='magenta', alpha=0.7)

for i in range(len(df)):
    plt.annotate(df['kebutuhan'][i], (df['tahun'][i], df['Propinsi'][i]), ha='center', va='bottom', fontsize=8)

plt.title('PLOT 1 - Kebutuhan tiap Propinsi setiap Tahunnya', fontsize=18)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Propinsi', fontsize=12)

plt.xticks(rotation=0, fontsize=8)
plt.grid(True, linestyle='--', alpha=1)

plt.tight_layout()
plt.show()
```

# ***Scatter Plot 2*** Kebutuhan Pertahun Propinsi Lampung

```python
df = pd.DataFrame(df)

provinsi_pilihan = 'LAMPUNG'
df_provinsi_pilihan = df.loc[df['Propinsi'] == provinsi_pilihan]

x = df_provinsi_pilihan['tahun']
y = df_provinsi_pilihan['kebutuhan']

plt.title(f'PLOT 2 - Statistik Kebutuhan Propinsi {provinsi_pilihan}')
plt.xlabel("Tahun")
plt.ylabel("Kebutuhan")

plt.scatter(x, y, color='magenta')

plt.xlim(min(x), max(x))
# Menampilkan data provinsi tertentu

plt.show()
```

# ***Scatter Plot 3*** Menampilkan Data dari File transaksi2

```python
df_transaksi = pd.read_csv("/content/transaksi2.csv", names=['date', 'id_product', 'quantity', 'id_penjualan'])
df_transaksi
```

```python
df_transaksi['date'] = pd.to_datetime(df_transaksi['date'])
df_transaksi['year'] = df_transaksi['date'].dt.year
df_transaksi['month'] = df_transaksi['date'].dt.month_name()
df_clean = df_transaksi.loc[df_transaksi['year'] == 2017]
df_clean

plt.title('PLOT 3 - Menampilkan Data dari File Transaksi2')
plt.scatter(df_clean['date'], df_clean['quantity'], color='magenta')

plt.xticks(rotation=90)
plt.xlabel('Date')
plt.ylabel('Quantity')

plt.show()
```

# ***Scatter Plot 4*** Menampilkan data pesanan setiap kota pada file SupplyKantor

```python
df = pd.read_csv('/content/SupplyKantor.csv')
```

```python
df
```

```python
x = df['Kota']
y = df['Jumlah']
plt.figure(figsize=(12, 10))
plt.scatter(x, y, color="magenta")

plt.xlabel('Kota')
plt.ylabel('Jumlah')

plt.title('PLOT 4 - Total Jumlah Pesanan dari Setiap Kota')

plt.show()
```

# ***Scatter Plot 5*** Menampilkan data jumlah pesanan setiap barang yang ada di Yogyakarta

```python
data_filtered = df[df['Kota']=='Yogyakarta']
data_filtered
```

```python
x = data_filtered['Jenis']
y = data_filtered['Jumlah']
plt.figure(figsize=(12, 10))
plt.scatter(x, y, color="magenta")
plt.xlabel('Jenis Barang')
plt.ylabel('Jumlah')

plt.title('PLOT 5 - Jumlah Pesanan dari Kota Yogyakarta')

plt.show()
```
